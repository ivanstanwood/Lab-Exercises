```python
from tkinter import *
import requests
import json
import random

root = Tk()
root.title("Trivia")
root.geometry("1000x800")

user_answer = StringVar()
user_answer.set("")

intext = requests.get("https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
json_file = intext.json()

score = 0
questions_right = 0

canvas = Canvas(root, width=800, height=250, bg="khaki1")
canvas.place(relx=.5, rely=.5,anchor= CENTER)
game_display = canvas.create_text(200, 75, text="", fill="black", font=('Arial 13 bold'))

def GetChoices():
    option_list = []
    global chosen_question
    chosen_question = random.choice(json_file)
    correct = chosen_question['correctAnswer']
    option_list.append(correct)
    loop = chosen_question['incorrectAnswers']
    for x in loop:
        option_list.append(str(x))
    random.shuffle(option_list)
    return option_list

def GetQuestion():
    question = chosen_question['question']
    return question

def CreateLabels(question, answer):
    Question_label = Label(canvas, text=question, font=('Arial 10 bold'))
    Question_label.place(x=50, y=10)
    y=50
    for i in answer:
        rad_btn = Radiobutton(canvas, text=i, variable=user_answer, command=Correct, value=i, font=('Arial 10 bold'))
        rad_btn.place(x=50, y=y)
        y+=20
    restart_but = Button(canvas, text="Next Question!", command=Restart)
    restart_but.place(x=50, y=200)
    score_lablab = Label(canvas, text="S", font="bold")
    score_lablab.place(x=700, y=100)
    questions_right_lablab = Label(canvas, text="QR", font="bold")
    questions_right_lablab.place(x=750, y=100)

def Correct():
    answer = user_answer.get()
    for i in json_file:
        if i["question"] == chosen_question["question"]:
            if i["correctAnswer"] == answer:
                labcor = Label(canvas, text="Correct!", font=("arial", 15, "bold"))
                labcor.place(x=50, y=160)
                global questions_right
                global score
                score += 10
                questions_right += 1
                score_lab = Label(canvas, text=score, font="bold")
                score_lab.place(x=700, y=150)
                questions_right_lab = Label(canvas, text=questions_right, font="bold")
                questions_right_lab.place(x=730, y=150)
            else:
                labinc = Label(canvas, text="Incorrect!", font=("arial", 15, "bold"))
                labinc.place(x=50, y=160)
                score -= 10

def Restart():
    canvas.destroy()    #I could not find anywhere how to replace items within the frame 
    

def main():
    score = 0
    questions_right = 0
    list_option = GetChoices()
    question_thing = GetQuestion()
    CreateLabels(question_thing, list_option)
    score_lab = Label(canvas, text=score, font="bold")
    score_lab.place(x=700, y=150)
    questions_right_lab = Label(canvas, text=questions_right, font="bold")
    questions_right_lab.place(x=750, y=150)

main()
root.mainloop()

#I was able to create a score/questions right system displayed in the lower right, but I could just not find a way to display a new
#question with a button. I tried finding .update() functions that worked with a frame but nothing online helped.
```
