```python
import csv
import requests
from bs4 import BeautifulSoup
import html5lib
import pandas as pd

csv_file = open("Lab4_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Name", "Release Year", "Rating"])

source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")
#print(soup.prettify())

big_section = soup.find("div", class_="lister-list")
#print(big_section.prettify())

year_section = big_section.find("span", class_="lister-item-year text-muted unbold")

for element in big_section.find_all("div", class_="lister-item-content"):
    name = element.a.text.strip()
    print(name)
    
    year = element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    print(year)
    
    maturity_rating = element.find("span", class_="certificate").text.strip()
    print(maturity_rating)
    
    csv_writer.writerow([name, year, maturity_rating])

sheet2 = pd.read_csv("Lab4_scrape.csv")
sheet2
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import html5lib
import pandas as pd

csv_file = open("Lab4_scrape2.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Name", "Description"])

page = 1

while page != 6:
    source = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup(source, "lxml")

    section = soup.find("ul", class_="search-results list-view")

    first_section = section.find("div", class_="item-description")
    first_name = first_section.span.a.text.strip()
    first_desc = first_section.find("span", class_="item-description-abstract").text.strip()
    link_sec = first_section.find("a")
    first_link = link_sec.get("href")

    csv_writer.writerow([first_name, first_desc, first_link])


    for element in section.find_all("li", class_="item"):   
        name_section = element.find("span", class_="item-description-title")
        name = name_section.find("a").text.strip()
        print(name)
        
        link_section = element.find("a")
        link = link_section.get("href")
        print(link)
        
        try:
            desc = element.find("span", class_="item-description-abstract").text.strip()
            print(desc)
        except AttributeError:
            desc = "No description for this item"
            print(desc)

        csv_writer.writerow([name, desc, link])
    page += 1

sheet2 = pd.read_csv("Lab4_scrape2.csv")
sheet2
```
