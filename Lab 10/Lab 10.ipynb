```python
from tkinter import *
import requests
import random
import time

root = Tk()
root.geometry("1200x200")
root.title("Quote Giver")

people_list = []
chosen_people_list = []


def NameList():
    url = "https://api.quotable.io/authors"
    intext = requests.get(url).json()
    results = intext["results"]
    for i in results:
        names = i['name']
        people_list.append(names)

def ListAssignment():
    for i in range(1,6):
        chosen_name = random.choice(people_list)
        if chosen_name in chosen_people_list:
            chosen_people_list.append(chosen_name)
            chosen_people_list.remove(chosen_name)
        else:
            chosen_people_list.append(chosen_name) 

Lab1 = Label(root, text="Choose from the drop-down menu to hear an individual's quote!").pack()

NameList()
ListAssignment()

def DisplayQuote():
    text_box.delete(0.0, END)
    text_box.insert(END, quote)
    
def Quote():
    person = value.get()
    result_label = Label(root, text=f"You've chosen {person}, here is a quote!").pack()
    url = "https://api.quotable.io/random"
    intext = requests.get(url).json()
    quote = intext['content']
    time.sleep(1)
    quote_label = Label(root, text=quote).pack()
    DisplayQuote()


value = StringVar()
value.set("Select a Famous Figure")
dropdown = OptionMenu(root, value, *chosen_people_list)
choose_button = Button(root, text="Click to Choose", command=Quote)

dropdown.pack()
choose_button.pack()

text_box = Text(root, font="Elephant", height=4, width=30, bg="#D1E1F3", padx=30, pady=50)

root.mainloop()

```

```python
from tkinter import *
import requests
import random

root = Tk()
root.geometry("700x200")
root.title("Weather Giver")
counter = 2

latitude = input("What is your latitude to 4 decimal places?")
longitude = input("What is you longitude to 4 decimal places?")

url = f"https://api.weather.gov/points/{latitude},{longitude}"
intext = requests.get(url).json()
#print(intext)
forecast_url = intext['properties']['forecast']
newrequest = requests.get(forecast_url)
json = newrequest.json()
print(json)
forecast = json['properties']['periods']

for i in forecast:
    if " Night" not in str(i['name']):
        day = i["name"]
        temp = i["temperature"]
        wind = i['windSpeed']
        Label1 = Label(root, text="Day")
        Label1.grid(row=0, column=0)
        Label1 = Label(root, text="Temp.")
        Label1.grid(row=1, column=0)
        Label1 = Label(root, text="Wind")
        Label1.grid(row=2, column=0)
        Label1 = Label(root, text=str(day))
        Label1.grid(row=0, column=counter)
        Label1 = Label(root, text=str(temp) + i["temperatureUnit"])
        Label1.grid(row=1, column=counter)
        Label1 = Label(root, text=wind)
        Label1.grid(row=2, column=counter)
    counter += 1

sbb = Scrollbar(root, orient=HORIZONTAL)
sbb.grid()

root.mainloop()
```
