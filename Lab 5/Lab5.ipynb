```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import selenium
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By                     
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import pandas as pd

PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service=PATH)  

csv_file = open("Lab5_scrape.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Description"])

website = "https://data.gov/"
driver.get(website)   

link = driver.find_element(By.XPATH, "//*[@id='menu-primary-navigation']/li[1]/a")
link.click()  

search = driver.find_element(By.CLASS_NAME, "search")
search.clear()
search.send_keys("climate change")
search.send_keys(Keys.RETURN)

try:
    section = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "module-content")))
    #print(section.text)
    
    things = section.find_elements(By.CLASS_NAME, "dataset-content")
    #print(names)
    
    for i in things:
        name = i.find_element(By.TAG_NAME, "a").text
        print(name)
        
        desc = i.find_element(By.CLASS_NAME, "notes").text
        print(desc)
        
        csv_writer.writerow([name, desc])
finally:
    driver.quit()

sheet2 = pd.read_csv("Lab5_scrape.csv")
sheet2
```
