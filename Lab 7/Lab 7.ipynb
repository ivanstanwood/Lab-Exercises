```python
import requests 
import json
import pandas as pd
import csv

lat = "40.6204"
lon = "-73.9600"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")

json_file = genius.json()
json_file

from matplotlib import pyplot as plt
import numpy as np

csv_file = open("Lab7_scrape1.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Day", "Temp", "Description"])

forecast_url = json_file['properties']['forecast']
newrequest = requests.get(forecast_url)
json = newrequest.json()
json

data_holder = json['properties']['periods']
#print(data_holder)

for i in data_holder:
    day = i["name"]
    print(i["name"])
    temp = i["temperature"]
    print(str(i["temperature"]) + i["temperatureUnit"])
    description = i["detailedForecast"]
    print(i["detailedForecast"])
    csv_writer.writerow([day, temp, description])
    
sheet = pd.read_csv("Lab7_scrape1.csv")
sheet

df=pd.DataFrame(sheet)

day = df["Day"]
temp = df["Temp"]

fig = plt.figure(figsize = (16, 8))
plt.bar(day[0:10], temp[0:10])


plt.xlabel("Day")
plt.ylabel("Temperature (F)")
plt.title("Week Forecast")

plt.show()
```

```python
import requests 
import json
import pandas as pd
import csv
from matplotlib import pyplot as plt
import numpy as np

csv_file = open("Lab7_scrape2.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Country Code", "Number of Holidays"])

countries_list = ["AU", "US", "DE", "US", "UA", "GB", "CN", "CH", "SE", "FR"]
year = "2023"
holiday_list = []
holiday_length_list = []

for i in countries_list:
    specific_holiday_list = []
    countryCode = i
    url = f"https://date.nager.at/api/v3/PublicHolidays/{year}/{countryCode}"
    print()
    print(countryCode)
    print()
    country_data = requests.get(url)
    json_file = country_data.json()
    json_file
    for i in json_file:
        holiday_name = i['name']
        print(holiday_name)
        holiday_list.append(holiday_name)
        specific_holiday_list.append(holiday_name)
    length_specific = len(specific_holiday_list)
    holiday_length_list.append(length_specific)
    csv_writer.writerow([countryCode, length_specific])

lengthof_list = len(holiday_list)
print()
print("Number of Holidays: " + str(lengthof_list))
print()

sheet = pd.read_csv("Lab7_scrape2.csv")
sheet

df=pd.DataFrame(sheet)

code = df['Country Code']
number = df['Number of Holidays']

fig = plt.figure(figsize = (10, 7))
plt.bar(code[0:10], number[0:10])


plt.xlabel("Country Code")
plt.ylabel("Number of Holidays")
plt.title("Number of Holidays in Country in 2023")

plt.show()
```
